```python
runs2process = dict(era5=[0], interim=[0])
subsets = ["pmc"]
```

```python
track_runs = {}
for dset in datasets:
    track_runs[dset] = {}
    for run_num in runs2process[dset]:
        track_runs[dset][f"run{run_num:03d}"] = TrackRun.from_archive(
            mypaths.procdir / f"{dset}_run{run_num:03d}_{period}.h5"
        )
```

Define what statistic to calculate according to bins and what title to use in the plot below.

```python
vrbl_dict = dict(
    #    gen_lys_dist_km=dict(bins=np.linspace(0, 1800, 7), title='Genesis-lysis distance [km]'),
    lifetime_h=dict(bins=np.linspace(0, 72, 13), title="Life time [h]"),  # [3, 9, 18, 36, 54, 72]
    total_dist_km=dict(bins=np.linspace(0, 1600, 9), title="Total track distance [km]"),
    average_speed=dict(bins=np.linspace(0, 90, 11), title="Propagation velocity [$km$ $h^{-1}$]"),
    max_vort=dict(bins=np.linspace(2, 7, 11), title="Maximum vorticity [$10^{-4}$ $s^{-1}$]"),
    area=dict(bins=np.linspace(50, 450, 9), title="Vorticity area diameter [km]"),
    slp=dict(bins=np.linspace(970, 1020, 11), title="Min SLP [hPa]"),
)
```

Extract statistics for every track in both datasets.

```python
track_summary = {}
for dset in datasets:
    track_summary[dset] = {}
    for subset in subsets:
        _data = {}
        for vrbl_name in progress_bar(vrbl_dict.keys()):
            _data[vrbl_name] = []
            for i, ct in track_runs[dset]["run000"][subset].groupby("track_idx"):
                if vrbl_name == "area":
                    _data[vrbl_name].append(2 * (getattr(ct, vrbl_name).mean() / np.pi) ** 0.5)
                elif vrbl_name == "slp":
                    _data[vrbl_name].append(ct.slp.min())
                else:
                    _data[vrbl_name].append(getattr(ct, vrbl_name))

        #     [[getattr(ct, vrbl_name) if vrbl_name != 'area' else 2*(getattr(ct, vrbl_name).mean()/np.pi)**0.5
        #                                            for vrbl_name in vrbl_dict.keys()]
        #                                           for i, ct in TR[subset].groupby('track_idx')]
        track_summary[dset][subset] = pd.DataFrame(_data)
        track_summary[dset][subset].max_vort *= 1e4
```

Plot results as histograms.

```python
plt_kw_list = [dict(color="C0", alpha=0.5), dict(color="C1", alpha=0.5)]
```

```python
# import string
iletters = iter(string.ascii_lowercase)

ymax = 50

ncols = 3
nrows = 2
fig, axs = plt.subplots(
    ncols=ncols, nrows=nrows, sharey="row", gridspec_kw=dict(hspace=0.3, wspace=0.2)
)

for ax, (vrbl_name, h_kw) in zip(axs.flat, vrbl_dict.items()):
    hist_kw = h_kw
    ttl = hist_kw["title"]
    bins = hist_kw["bins"]

    for dset, plt_kw in zip(datasets, plt_kw_list):
        clipped_data = np.clip(track_summary[dset][subset][vrbl_name], bins[0], bins[-1])
        clipped_data.hist(
            ax=ax,
            label=aliases[dset],
            weights=np.ones(clipped_data.shape) * 100 / (clipped_data.shape[0]),  # nyr
            bins=bins,
            edgecolor=plt_kw["color"],
            **plt_kw,
        )
        med_val = np.median(track_summary[dset][subset][vrbl_name])
        ax.plot(
            med_val,
            0.9 * ymax,
            alpha=plt_kw.get("alpha", 1),
            mfc=plt_kw["color"],
            marker="o",
            ms=8,
            mec="k",
            mew=0.75,
            zorder=100,
        )
    ax.grid(False)
    ax.set_xlim(bins[0], bins[-1])
    ax.set_xticks(bins[1:-1:2])
    if np.equal(np.mod(bins, 1), 0).all():
        bin_labels = [f"{int(x)}" for x in bins]
    else:
        bin_labels = [f"{np.round(x, 1)}" for x in bins]
    # bin_labels[0] = ""  # f"$<{bin_labels[1]}$"
    # bin_labels[-1] = ""  # f"$>{bin_labels[-2]}$"
    bin_labels = bin_labels[1:-1:2]
    ax.set_xticklabels(bin_labels)
    ax.set_ylim(0, ymax)
    #     ax.set_yscale('log')
    ax.set_title(f"({next(iletters)}) {ttl}", loc="left", fontsize="large")
    #     ax.add_artist(AnchoredText(f'({next(iletters)})', loc=1, frameon=True))

    for spine in ax.spines.values():
        if spine.spine_type in ["top", "right"]:
            spine.set_linewidth(0)
        else:
            spine.set_linewidth(1)
            spine.set_position(("axes", -0.02))
            spine.set_position(("axes", -0.02))
#     ax.tick_params(axis="x", labelsize="small")

for axrow in axs:
    axrow[0].set_ylabel("Relative frequency [%]")
axs.flatten()[0].legend(
    fontsize="x-large", ncol=4, loc=2, bbox_to_anchor=(-0.05, 1.35), frameon=False
)
plt.close()  # Show the figure in a separate cell
```
